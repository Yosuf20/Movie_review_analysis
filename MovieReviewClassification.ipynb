```python
df = pd.read_csv("final.csv", index_col=0)
```

```python
df
```

```python
import nltk
nltk.download('punkt_tab')
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for review in df["clean_text"]:
  raw_sent = sent_tokenize(review)
  for w in raw_sent:
    story.append(simple_preprocess(w))

```

```python
story
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

Y = encoder.fit_transform(df['sentiment'])
X = df['clean_text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(
    max_features=10000,
    min_df=5,
    max_df=0.8
)
tf_train = tf.fit_transform(X_train)
tf_val = tf.transform(X_val)
```

```python
tf_train[1].shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score


rf = RandomForestClassifier()
rf.fit(tf_train, Y_train)
y_pred = rf.predict(tf_val)
print(accuracy_score(Y_val, y_pred))
print(confusion_matrix(Y_val, y_pred))
print(precision_score(Y_val, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()

LR.fit(tf_train, Y_train)

y_pred = LR.predict(tf_val)
print(accuracy_score(Y_val, y_pred))
print(confusion_matrix(Y_val, y_pred))
print(precision_score(Y_val, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB()

MNB.fit(tf_train, Y_train)
y_pred = MNB.predict(tf_val)

print(accuracy_score(Y_val, y_pred))
print(confusion_matrix(Y_val, y_pred))
print(precision_score(Y_val, y_pred))
```
