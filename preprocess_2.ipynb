```python
df["review"] = df["review"].str.lower()
```

```python
import re

def remove_html(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

df["review"] = df["review"].apply(remove_html)
```

```python
import string

punc = string.punctuation

def remove_punc(text):
    for char in punc:
        text = text.replace(char, '')
    return text

df["review"] = df["review"].apply(remove_punc)
```

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

df["review"] = df["review"].apply(remove_url)
```

```python
def chat_treat(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_word:
            new_text.append(chat_word[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

df["review"] = df["review"].apply(chat_treat)
```

Spelling Correction


```python
from textblob import TextBlob

def spell_check(text):
    spell = TextBlob(text)
    return spell.correct().string

df["review"] = df["review"].apply(spell_check)
```

```python

def normalize_words(text):
    # replace repeated characters (more than 2 → 2)
    return re.sub(r'(.)\1{2,}', r'\1\1', text)

df["review"] = df["review"].apply(normalize_words)
```

Stop Words Removal


```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))


def removewords(text):
    new = []
    for w in text.split():
        if w not in stop_words:
            new.append(w)
    return " ".join(new)

df["review"] = df["review"].apply(removewords)
```

Tokenization


```python
from nltk.tokenize import word_tokenize

df["token"] = (df["review"]).apply(word_tokenize)
```

lemmitization

```python
from nltk.stem import WordNetLemmatizer

lem = WordNetLemmatizer()

def word_lem(text):
    new = []
    for w in text:
        new.append(lem.lemmatize(w,pos='v'))
    return " ".join(new)

df["token"] = df["token"].apply(word_lem)
```

```python
import re

def clean_text(text):
    # remove special/control characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    
    # remove numbers
    text = re.sub(r'\d+', '', text)
    
    # remove punctuation (optional)
    text = re.sub(r'[^\w\s]', '', text)
    
    # lowercase
    text = text.lower()
    
    return text

df["clean_text"] = df["token"].apply(clean_text)
```

```python
vocab = set()

for text in df["clean_text"]:
    vocab.update(text.split())

vocab = sorted(vocab)
print(len(vocab))
```
